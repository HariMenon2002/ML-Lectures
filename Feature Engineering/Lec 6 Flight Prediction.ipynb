```python
#Now we have to handle the cols one by one
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]

print(df['Date'])

#but the above 3 are still object type
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
#Arrival Time- let us first remove the date from that and keep only time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
#Arrival Time- we will have to split the time over :

df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)

df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
#Same thing can be done with Dep_Time
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]

df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)

df.drop('Dep_Time',axis=1,inplace=True)
df.head(2)
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
#so that nan value has to be replaced with 1 stop
#also we will do ordinal encoding here as no of stops increases, flight price also increase
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head(2)
```

```python
#as i have source and destn, i dont need route
df.drop('Route',axis=1,inplace=True)
```

```python
#To handle Duration
df['Duration_hours']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_mins']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df.drop('Duration',axis=1,inplace=True)

```

```python
df.head(2)
```

```python
df['Duration_mins'].isnull().sum()  
```

```python
#Thus so many null values in mins has to be converted to 0
df['Duration_mins']=df['Duration_mins'].fillna(0)
df['Duration_mins']
```
